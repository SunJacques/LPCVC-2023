```python
import matplotlib.pyplot as plt
import os
from PIL import Image
import numpy as np
```

```python
#Load Image from the dataset

folder_path = "../dataset/train/GT/"
folder_path_label = "../dataset/train/IMG/"
image_files = os.listdir(folder_path)
image_files_label = os.listdir(folder_path_label)
```

```python
num_images_to_display = 10

fig = plt.figure(figsize=(16, 8))

for i in range(num_images_to_display):
    image_file = os.path.join(folder_path_label, image_files_label[i])
    image = np.asarray(Image.open(image_file))

    ax = fig.add_subplot(2, 5, i + 1)
    ax.imshow(image, vmin=0, vmax=15, cmap='jet', aspect='auto')
    ax.axis("off")

plt.show()
```
